```python
# 4️⃣ Load SAM
from segment_anything import sam_model_registry, SamPredictor
import cv2
import numpy as np
import matplotlib.pyplot as plt

sam = sam_model_registry["vit_b"](checkpoint="sam_vit_b_01ec64.pth")
predictor = SamPredictor(sam)
```

```python
# 5️⃣ Function to run SAM segmentation and save results
def run_sam_segmentation(img_path, save_path):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    predictor.set_image(img)

    # Example: one foreground click at center
    h, w, _ = img.shape
    input_points = np.array([[w//2, h//2]])
    input_labels = np.array([1])

    masks, scores, _ = predictor.predict(point_coords=input_points,
                                         point_labels=input_labels,
                                         multimask_output=False)

    # Visualize Before/After
    plt.figure(figsize=(8,4))
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.title("Before")
    plt.subplot(1,2,2)
    plt.imshow(img)
    plt.imshow(masks[0], alpha=0.5)
    plt.title("After")
    plt.show()

    # Save mask overlay
    mask_save_path = os.path.join(save_path, os.path.basename(img_path).split('.')[0]+"_after.png")
    plt.imsave(mask_save_path, masks[0], cmap='gray')
```

```python
import os

# Input folder (read-only)
img_folder = "/kaggle/input/chest-ctscan-images"

# Output folder (writeable)
result_folder = "/content/results/before_after"
os.makedirs(result_folder, exist_ok=True)

# Run SAM segmentation on all images
for root, dirs, files_list in os.walk(img_folder):
    for f in files_list:
        if f.lower().endswith(('.png','.jpg','.jpeg')):
            run_sam_segmentation(os.path.join(root,f), result_folder)
```

```python
# 7️⃣ Compute IoU if you have ground truth masks
def iou(pred_mask, true_mask):
    intersection = np.logical_and(pred_mask, true_mask).sum()
    union = np.logical_or(pred_mask, true_mask).sum()
    return intersection / union
```

```python
# 8️⃣ Classification with transfer learning (cancer vs normal)
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms, models, datasets
from torch.utils.data import DataLoader

# Transform & dataset
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor()
])

dataset = datasets.ImageFolder("/kaggle/input/chest-ctscan-images", transform=transform)  # assumes subfolders per class
loader = DataLoader(dataset, batch_size=16, shuffle=True)
```

```python
# Model setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet18(pretrained=True)
model.fc = nn.Linear(model.fc.in_features, len(dataset.classes))
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
# Quick 1-epoch training loop (demo)
for images, labels in loader:
    images, labels = images.to(device), labels.to(device)
    optimizer.zero_grad()
    outputs = model(images)
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()
print("Training step completed (demo, 1 epoch)")
```
